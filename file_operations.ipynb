Using your knowledge of Python and Matplotlib write your code to transform and graph the data below.

After your code create a new Markdown block to describe your analysis

````python
import matplotlib.pyplot as plt
num_Fl=0
num_Ny=0
num_Mo=0#this part is used to stored all of states which will be in chart
num_Ok=0
num_Mn=0

with open ("states.txt","r") as listt:
    for x in listt:
        if x == "FL\n":
          num_Fl+=1
        elif x == "NY\n":#this part is counting states and saving in valuble
          num_Ny+=1
        elif x == "MO\n":
          num_Mo+=1
        elif x == "OK\n":
          num_Ok+=1
        elif x == "MN\n":
          num_Mn+=1


labels='FL',"NY","MO","MN","OK"
sizes=num_Mo,num_Fl,num_Mn,num_Ok,num_Ny
ax=plt.subplot()#this part is printing pie chart

ax.pie(sizes,labels=labels)
plt.title("states")
````
