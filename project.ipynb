```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("NYCLIB.csv")
df.head()

#this part of ipynb file is used to import stuff which will be needed for later use such as csv file,matplotlib and pandas.
```

```python
graph1 = df.groupby('Boro/Central Library')['CIRCULATION Young Adult'].sum()#colects data from csv file about circulation of Young adults.
```

```python
plt.style.use('bmh')

bar_colors = ['tab:red', 'tab:blue', 'tab:pink', 'tab:orange','tab:purple','tab:green']
graph1.plot(kind="bar",ylabel='CIRCULATION Young Adult',color=bar_colors)

plt.xticks(rotation=65)#xticks rotates names of x values in ordert to make it easy to read

```
