```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("titanic.csv")
df.head()

#this block is importing block with only imports and read code
```

```python
ploty = df.groupby(['Sex','Survived']).Age.mean()#this shows average age of person on ship
print(ploty)
ploty.plot(kind='bar')

```

```python
totalu=df.groupby("Pclass").Survived.mean()
totalu.plot(kind="bar",ylabel='procent')
```

```python
plote = df.groupby(['Age','SibSp']).Survived.mean()#
plote.plot(kind='hist',ylabel='Age',xlabel='Survived')
```
